```python
#import biblioteka koje smo instalirali
import os #operativni sistem
import numpy as np #za matematicke operacije
import pandas as pd #za tabele
import mne #za EEGpodatke
from scipy.signal import butter,filtfilt #obrada signala
from scipy import signal 
import matplotlib.pyplot as plt #za grafike
import requests #za preuzimanje fajlova sa interneta
from pathlib import Path #kreiranje foldera i manipulacija putanjama
import warnings #za ignorisanje upozorenja zato dole stavljamo ignore
warnings.filterwarnings("ignore")


#Konstante
PROJECT_ROOT=Path("C:/Users/Cofiliks/NeuronskeMreze_Projekat") #glavna putanja projekta
DATA_RAW_DIR=PROJECT_ROOT/ "data" / "raw" #gde se skidaju EDF podaci
DATA_PROCESSED_DIR=PROJECT_ROOT / "data" / "processed" #gde se cuvaju obradjeni podaci

#parametri za obradu
SAMPLING_RATE=256 #CHB se meri na 256 uzoraka po sekundi
FILTER_LOW=0.5 #band-pass(propusta samo odredjeni opseg frekvencija a sve ostale uklanja) 
FILTER_HIGH=100 #filter cuva samo frekvencije od 0.5-100HZ epilepticne
WINDOW_LENGTH=3 #velicina prozora u sekundama tj svaki segment traje 3 sekunde
OVERLAP=0.5#procenat preklapanje izmejdu prozora, 50%
PATIENTS=['chb01','chb02','chb03','chb04','chb05'] #lista pacijenata koje cemo obraditi,skidamo podatke za 5 pacijenata

#kreiramo foldere ako ne postoje
#mkdir-make directory i kazemo da napravi i sve roditeljske foldere 
DATA_RAW_DIR.mkdir(parents=True,exist_ok=True)
DATA_PROCESSED_DIR.mkdir(parents=True,exist_ok=True)

#ispis poruka
print(f"✅ Konstante postavljene")
print(f"   Sampling rate: {SAMPLING_RATE} Hz")
print(f"   Filter band: {FILTER_LOW}-{FILTER_HIGH} Hz")
print(f"   Window: {WINDOW_LENGTH}s sa {OVERLAP*100}% overlap")
```

```python
#Prelazimo na skidanje chb-mit dataseta
#Definisemo funkciju koja ce preuzimati fajlove
#parametri: lista pacijenata i direktorijum gde ce se cuvati skidani podaci
def download_chb_mit_dataset(patient_ids,download_dir):
    """
    Download CHB-MIT EEG dataset za specificne pacijente.
    Physionet je javno dostupan dataset.
    """
    #url gde se nalaze podaci na netu
    base_url="https://physionet.org/files/chbmit/1.0.0"

    #prolazimo kroz listu pacijenata, za svakog pravimo folder i skidamo njegove podatke
    for patient_id in patient_ids:
        patient_dir=download_dir / patient_id
        patient_dir.mkdir(exist_ok=True)

        #Download .edf fajlove za ovog pacijenta
        print(f"\n📥 Skidanje podataka za {patient_id}...")

        #CHB-MIT ima nekoliko .edf fajlova po pacijentu
        #skinucemo samo prva 2-3 recenja(EEG recordings/sesije) za brze testiranje
        for recording_num in range(1,4): #1,2,3
            filename=f"{patient_id}_{recording_num:02d}.edf" #chb01.edf, chb01_02.edf, chb01_03.edf 
            file_url=f"{base_url}/{patient_id}/{filename}" #url na netu/chb01/chb01_01.edf
            file_path=patient_dir / filename #gde ce biti spreman lokalno

            if file_path.exists():
                print(f"   ✓ {filename} već postoji")
                continue
            
             #prvo preuzimamo fajl sa urla,ako trja duze od 30 sekundi,prekidamo
            #ako je status code 200, znaci da je fajl pronadjen i usoesno preuzet
            #zatim otvaramo fajl za pisanje u binarnom modu (wb)
            #upisujemo preuzete podatke u fajl i sledi poruka o uspehu
            #ako nije status code 200, fajl nije pronadjen i ispisujemo poruku za to 
            #except nam sluzi ako se desi neka greska tokom skidanja, da uhvatimo tu gresku i ispisemo poruku o tome sta se desilo
            #zatim se ispisuje poruka o gotovom skidanju kada je sve gotovo
            
            try:
                response=requests.get(file_url,timeout=30)
                if(response.status_code==200):
                    with open(file_path,'wb') as f:
                        f.write(response.content)
                    print(f"Preuzet:{filename}")
                else:    
                    print(f"Nije pronaden:{filename}")
            except Exception as e:
                print(f"Greska pri skidanju {filename}:{str(e)}")

    print(f"\nDownload zavrsen!")

#pokrecemo download
#zovemo funkciju sa nasim parametrima,lista pacijenata i direktorijum gde ce se cuvati skidani podaci
download_chb_mit_dataset(PATIENTS,DATA_RAW_DIR)
```
