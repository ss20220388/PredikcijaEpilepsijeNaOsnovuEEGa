```python
# Preuzimanje CHB-MIT dataseta sa seizure anotacijama

def download_chb_mit_dataset(patient_ids, download_dir):
    """
    Download CHB-MIT EEG dataset za specificne pacijente
    + summary fajlove sa seizure anotacijama
    """

    base_url = "https://physionet.org/files/chbmit/1.0.0"

    for patient_id in patient_ids:

        patient_dir = download_dir / patient_id
        patient_dir.mkdir(parents=True, exist_ok=True)

        print(f"\n📥 Skidanje podataka za {patient_id}...")

        # ===============================
        # 1. Skidanje summary fajla
        # ===============================

        summary_filename = f"{patient_id}-summary.txt"
        summary_url = f"{base_url}/{patient_id}/{summary_filename}"
        summary_path = patient_dir / summary_filename

        if summary_path.exists():
            print(f"   ✓ {summary_filename} već postoji")
        else:
            try:
                response = requests.get(summary_url, timeout=30)

                if response.status_code == 200:
                    with open(summary_path, "wb") as f:
                        f.write(response.content)

                    print(f"   ✓ Preuzet {summary_filename}")
                else:
                    print(f"   ❌ Nije pronađen {summary_filename}")

            except Exception as e:
                print(f"   ❌ Greška pri skidanju {summary_filename}: {str(e)}")

        # ===============================
        # 2. Skidanje EDF fajlova
        # ===============================

        for recording_num in range(1, 4):

            filename = f"{patient_id}_{recording_num:02d}.edf"

            file_url = f"{base_url}/{patient_id}/{filename}"

            file_path = patient_dir / filename

            if file_path.exists():
                print(f"   ✓ {filename} već postoji")
                continue

            try:
                response = requests.get(file_url, timeout=30)

                if response.status_code == 200:

                    with open(file_path, "wb") as f:
                        f.write(response.content)

                    print(f"   ✓ Preuzet {filename}")

                else:
                    print(f"   ❌ Nije pronađen {filename}")

            except Exception as e:
                print(f"   ❌ Greška pri skidanju {filename}: {str(e)}")

    print("\n✅ Download završen!")
```

```python
#Ucitavanje i ekspolacija EDF fajlova
#kreiramo funkciju koja ce ucitati jedan EDF fajl
#ona prima paramtar edf_file tj putoanju do edf fajla i vraca informacije o njemu
def load_and_explore_edf(edf_file):
    """
    Ucitaj EDF fajl i prikazi osnovne informacije
    """
    #uvek gledamo da stavimo try blok kada radimo sa fajlovima da bismo uhvtatili greske 
    #mne.io.read_raw_edf je funkcija koja cita edf fajl i stavljamo da preload ucita sve podatke u memoriju
    #raw je Python objekat koji sadrzi sve EEG podatke
    try:
        raw =mne.io.read_raw_edf(edf_file,preload=True)

        info={
            'file': edf_file.name, #naziv fajla
            'sampling_rate': raw.info['sfreq'], #kolikon puta po sekundi merimo signal
            'duration_seconds':raw.times[-1], #koliko dugo traje snimanje
            'n_channels':len(raw.ch_names), #broj kanala tj. senzora koji su koristeni za merenje
            'channels':raw.ch_names, #lista imena kanala
            'n_samples':raw.n_times #broj uzoraka
        }
        return raw,info
    except Exception as e:
        print(f"Greska pri ucitavanju{edf_file}:{str(e)}")
        return None,None
    
#Pronalazimo sve edf fajlove i sortiramo ih alfabetski
edf_files=sorted(DATA_RAW_DIR.glob("**/*.edf"))
print(f"Pronadjeno EDF fajlova:{len(edf_files)}")

#ucitamo i prikazujemo informacije za prvi fajl
if edf_files:
    #dve varijable koje dobijamo iz funkcije load_and_exploare_edf
    raw_example,info_example=load_and_explore_edf(edf_files[0])
    if(raw_example):
        print(f"Primer fajla: {info_example['file']}")
        print(f"Sampling rate: {info_example['sampling_rate']} Hz")
        print(f"Trajanje: {info_example['duration_seconds']:.2f} sekundi")
        print(f"Broj kanala: {info_example['n_channels']}")
        print(f"Kanali: {info_example['channels']}")
        print(f"Broj uzoraka: {info_example['n_samples']}")
    
else:
    print("Nema pronadjenih EDF fajlova za ucitavanje.")
```

```python
# ════════════════════════════════════════════════════════════════════════════
# 🔄 DATA AUGMENTATION: 24 → ~100 Pseudo-Pacijenata sa Seizure Labelama
# ════════════════════════════════════════════════════════════════════════════

import re

print("\n" + "="*80)
print("🔄 FAZA 1B: DATA AUGMENTATION + SEIZURE LABELING")
print("="*80 + "\n")

# ─────────────────────────────────────────────────────────────────────────
# 📖 ČITAJ SEIZURE ANOTACIJE IZ .txt FAJLOVA
# ─────────────────────────────────────────────────────────────────────────
def load_seizure_annotations(patient_id):
    """
    Učitaj seizure anotacije iz CHB-MIT summary fajla.

    Vraća:
    {
        'chb01_03.edf': [
            {'start': 2996, 'end': 3036}
        ],
        ...
    }
    """

    annotations = {}

    # Lokacija summary fajla
    txt_file = DATA_RAW_DIR / patient_id / f"{patient_id}-summary.txt"

    if not txt_file.exists():
        print(f"⚠️ Nema summary fajla za {patient_id}")
        return annotations

    current_file = None

    try:
        with open(txt_file, "r", encoding="utf-8", errors="ignore") as f:

            for line in f:
                line = line.strip()

                # File Name: chb01_03.edf
                if "File Name:" in line:

                    current_file = line.split("File Name:")[-1].strip()

                    if current_file not in annotations:
                        annotations[current_file] = []

                # Seizure Start Time: 2996 seconds
                elif "Seizure Start Time" in line:

                    match = re.search(r"(\d+)", line)

                    if match and current_file:

                        annotations[current_file].append({
                            "start": int(match.group(1)),
                            "end": None
                        })

                # Seizure End Time: 3036 seconds
                elif "Seizure End Time" in line:

                    match = re.search(r"(\d+)", line)

                    if (
                        match
                        and current_file
                        and current_file in annotations
                        and len(annotations[current_file]) > 0
                    ):

                        annotations[current_file][-1]["end"] = int(match.group(1))

    except Exception as e:
        print(f"❌ Greška pri čitanju {txt_file}: {e}")

    return annotations

# ─────────────────────────────────────────────────────────────────────────
# 🎲 AUGMENTACIJSKE TEHNIKE
# ─────────────────────────────────────────────────────────────────────────

def augment_edf_signal(signal, n_augmentations=3, seed=42):
    """
    Kreiraj augmentovane verzije signala
    Tehnike: noise, scaling, time-shift, smoothing
    """
    np.random.seed(seed)
    augmented_signals = []
    n_channels, n_samples = signal.shape
    
    for aug_idx in range(n_augmentations):
        aug_signal = signal.copy().astype(np.float64)
        
        # 1️⃣ Noise injection (3-7%)
        noise_level = np.random.uniform(0.03, 0.07)
        noise = np.random.normal(0, noise_level * np.std(aug_signal), aug_signal.shape)
        aug_signal = aug_signal + noise
        
        # 2️⃣ Amplitude scaling (0.9-1.1x)
        scale = np.random.uniform(0.9, 1.1)
        aug_signal = aug_signal * scale
        
        # 3️⃣ Time shifting (svaki drugi)
        if aug_idx % 2 == 0:
            shift = np.random.randint(-500, 500)
            aug_signal = np.roll(aug_signal, shift, axis=1)
        
        # 4️⃣ Gaussian smoothing (svaki treći)
        if aug_idx % 3 == 0:
            from scipy.ndimage import gaussian_filter1d
            aug_signal = gaussian_filter1d(aug_signal, sigma=0.2, axis=1)
        
        augmented_signals.append(aug_signal.astype(np.float32))
    
    return augmented_signals

# ─────────────────────────────────────────────────────────────────────────
# 🏭 KREIRAJ PSEUDO-PACIJENTE SA AUGMENTACIJOM
# ─────────────────────────────────────────────────────────────────────────

def create_augmented_dataset(original_patients, raw_dir, n_aug_per_patient=3):
    """
    Kreiraj augmentovane pseudo-pacijente
    """
    all_patients = []
    augmentation_log = []
    
    for idx, patient_id in enumerate(original_patients, 1):
        patient_dir = raw_dir / patient_id
        
        if not patient_dir.exists():
            print(f"[{idx}/{len(original_patients)}] ⚠️  {patient_id} - ne postoji folder")
            continue
        
        # Pronađi sve EDF fajlove
        edf_files = sorted(patient_dir.glob(f"{patient_id}_*.edf"))
        
        if not edf_files:
            print(f"[{idx}/{len(original_patients)}] ⚠️  {patient_id} - nema EDF fajlova")
            continue
        
        print(f"[{idx}/{len(original_patients)}] 🔄 {patient_id}...", end="", flush=True)
        
        try:
            # Učitaj prvi EDF
            raw = mne.io.read_raw_edf(edf_files[0], preload=True, verbose=False)
            signal_data = raw.get_data()
            fs = int(raw.info['sfreq'])
            ch_names = raw.ch_names
            info = mne.create_info(ch_names=ch_names, sfreq=fs, ch_types='eeg')
            
            # Originalni pacijent
            all_patients.append(patient_id)
            
            # Kreiraj augmentovane verzije
            augmented_signals = augment_edf_signal(signal_data, n_augmentations=n_aug_per_patient)
            
            for aug_idx, aug_signal in enumerate(augmented_signals, 1):
                pseudo_id = f"{patient_id}_aug{aug_idx}"
                
                # Kreiraj folder
                pseudo_dir = raw_dir / pseudo_id
                pseudo_dir.mkdir(exist_ok=True)
                
                # Spremi kao EDF
                pseudo_raw = mne.io.RawArray(aug_signal, info, verbose=False)
                pseudo_raw.save(
                    pseudo_dir / f"{pseudo_id}_01.raw.fif",
                    overwrite=True,
                    verbose=False
                )
                
                all_patients.append(pseudo_id)
            
            print(f" ✅ (+{n_aug_per_patient})")
            augmentation_log.append({
                'original': patient_id,
                'n_augmented': n_aug_per_patient,
                'total_edf_files': len(edf_files)
            })
            
        except Exception as e:
            print(f" ❌ Greška: {str(e)}")
            continue
    
    return all_patients, augmentation_log

# ─────────────────────────────────────────────────────────────────────────
# 🚀 PRIMIJENI AUGMENTACIJU
# ─────────────────────────────────────────────────────────────────────────

print(f"📝 Originalnih pacijenata: {len(PATIENTS)}")
print(f"📊 Augmentacija po pacijentu: {N_AUGMENTATIONS_PER_PATIENT}\n")

ALL_PATIENTS, aug_log = create_augmented_dataset(
    PATIENTS,
    DATA_RAW_DIR,
    n_aug_per_patient=N_AUGMENTATIONS_PER_PATIENT
)

print(f"\n{'='*80}")
print(f"✅ AUGMENTACIJA ZAVRŠENA!")
print(f"{'='*80}\n")

print(f"📋 FINALNA STATISTIKA:")
print(f"   Originalnih pacijenata: {len(PATIENTS)}")
print(f"   Augmentovanih pseudo-pacijenata: {len(ALL_PATIENTS) - len(PATIENTS)}")
print(f"   UKUPNO: {len(ALL_PATIENTS)} pacijenata ✅\n")

print(f"   Primjeri:")
print(f"   - {ALL_PATIENTS[0]} (original)")
if len(ALL_PATIENTS) > len(PATIENTS):
    print(f"   - {ALL_PATIENTS[len(PATIENTS)]} (first augmented)")
print(f"   - {ALL_PATIENTS[-1]} (last)\n")
```

```python
#Band-pass filtriranje signala tj. cuvamo samo frekvencije u odredjenom opsegu u nasem slucaju od 0.5 do 100Hz
#Blokira sve ispod 0.5Hz i iznad 100Hz
def apply_bandpass_filter(signal_data,fs,lowcut,highcut,order=4):
    """
    Primeni band-pass Butterworth filter na signal
    Paramteri:
    -signal_data:NumPy array sa signala 
    -fs:sampling rate(Hz) u nasem slucaju 256Hz
    -lowcut:donja frekvencija (Hz) (0.5)
    -highcut:gornja frekvencija (Hz) (100)
    -order: red filtera(4 je dobar kompromis izmedju permonansa i efikasnosti)
    """

    nyquist=fs/2 #maksimalna frekvencija koju mozemo izmeriti znaci 256/2=128Hz
    low=0.001 #normalizovana donja frekvencija lowcut/nyquist
    high=0.1 #normalizovana gornja frekvencija highcut/nyquist

    #Kreiraj Butterworth filter
    #butter kreira parametre za Butterworth filter
    #order nam govori koliko strm treba da bude filter
    #low i high normlaizovane frekvencije koje zelimo da cuvamo
    #btype=band znaci da zelimo band-pass filter
    #b i a su koeficijenti diferencijalne jednacine filtera- koriste se za filtera na signal 
    b,a=butter(order,[low,high],btype='band')

    #primeni filter na signal
    filtered_signal=filtfilt(b,a,signal_data)
    return filtered_signal

#Test filtera na primeru
if raw_example:
    #uzimamo prvi kanal i primenjujemo filter
    test_signal=raw_example.get_data()[0,:5000] #uzimamo samo prvih 5000 uzoraka
    filtered_test=apply_bandpass_filter(test_signal,SAMPLING_RATE,FILTER_LOW,FILTER_HIGH)

    print(f"Filter primjenjen")
    print(f"Originalni signal - Min: {test_signal.min():.2f}, Max: {test_signal.max():.2f}")
    print(f"Filtrirani signal - Min: {filtered_test.min():.2f}, Max: {filtered_test.max():.2f}")

    #Vizualizacija grafika originalnog i filtriranog signala
    fig, (ax1,ax2)=plt.subplots(2,1,figsize=(12,6))

    time_axis=np.arange(len(test_signal))/SAMPLING_RATE

    #gornji graf - originalni signal
    ax1.plot(time_axis,test_signal,linewidth=0.8)
    ax1.set_title("Originalni signal")
    ax1.set_ylabel("Amplituda u mikrovoltima")
    ax1.grid(True,alpha=0.3)

    #donji graf - filtrirani signal
    ax2.plot(time_axis,filtered_test,linewidth=0.8,color='green')
    ax2.set_title("Filtrirani signal")
    ax2.set_ylabel("Amplituda u mikrovoltima")
    ax2.set_xlabel("Vreme u sekundama")
    ax2.grid(True,alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("Grafik prikazan!")
```

```python
#Segmentacija vremenskih serija tj.deljenje dugog signala na manje fiksne delove(prozore) sa preklapanjem
def segment_signal(signal_data,fs,window_length_sec,overlap_ratio):
    """
    Podeli signal na prozore sa preklapanjem
    Parametri:
    -signal_data:Numpy array sa signalom (samples,) ili (channels, samples)
    -fs:sampling rate(Hz)
    -window_length:duzina prozora u sekundama (npr. 3s)
    -overlap_ratio:procenat preklapanja izmedju prozora (npr. 0.5 za 50%)

    Vraca:
    -segments:Numpy array sa dimenzijama (n_segments, n_channels, n_samples_per_window)
    """

    #broj uzoraka po prozoru i pomak izmedju prozora
    window_samples=int(window_length_sec*fs)
    step_samples=int(window_samples*(1-overlap_ratio))

    #Odredi broj segmenata
    if signal_data.ndim==1:
        signal_data=signal_data[np.newaxis,:]
    
    n_channels,n_samples=signal_data.shape
    n_segments=(n_samples-window_samples)//step_samples + 1 #broj segmenata koji mozemo dobiti iz signala sa datim parametrima


    #kreiraj segmente tj praznu matricu za segmente sa dimenzijama (n_segments, n_channels, window_samples)
    segments=np.zeros((n_segments,n_channels,window_samples))
    #popunjavamo segmente prolazeci kroz signal 
    for i in range(n_segments):
        start_idx=i*step_samples #pocetak prozora
        end_indx=start_idx+window_samples #kraj prozora
        segments[i]=signal_data[:,start_idx:end_indx]
    return segments

#Test segmentacije
print("Testiranje segmentacije...")
if raw_example:
    test_signal=raw_example.get_data()[:,:SAMPLING_RATE*60]
    segments=segment_signal(test_signal,SAMPLING_RATE,WINDOW_LENGTH,OVERLAP)

    print(f"Originalni signal: {test_signal.shape}")
    print(f"Broj segmenata: {segments.shape[0]}")
    print(f"Dimenzije svakog segmenta: ({segments.shape[1]} kanala, {segments.shape[2]} uzoraka)")
    print(f"Trajanje svakog prozora: {segments.shape[2] / SAMPLING_RATE:.2f} sekundi")
    print(f"Segmentacija uspješna!")
```

```python
#Standardizacija
def normalize_segments(segments):
    """
    Normalizuj segmente koriscenjem Z-score normalizacije
    (X-mean)/std
    """

    #kreiramo praznu matricu za normalizovane vrednosti 
    normalized=np.zeros_like(segments,dtype=np.float32)

    for i in range(segments.shape[0]): #za svaki segment
        for j in range(segments.shape[1]): #za svaki kanal
            signal_segment=segments[i,j,:]

            #racunamo mean i std za ovaj segment i kanal tj srednju vrednost i standardnu devijaciju
            mean=np.mean(signal_segment)
            std=np.std(signal_segment)

            #Zaobidji deljenje sa 0 ako je std jako mali, samo oduzmi mean
            if std>1e-10:
                normalized[i,j,:]=(signal_segment-mean)/std
            else:
                normalized[i,j,:]=signal_segment-mean

    return normalized

#Test normalizacije
print("Testiranje normalizacije...")
test_segments=segment_signal(test_signal,SAMPLING_RATE,WINDOW_LENGTH,OVERLAP)
normalized_segments=normalize_segments(test_segments)


print(f"   Prije normalizacije:")
print(f"      Mean: {test_segments.mean():.4f}")
print(f"      Std: {test_segments.std():.4f}")
print(f"      Min: {test_segments.min():.4f}")
print(f"      Max: {test_segments.max():.4f}")

print(f"   Poslije normalizacije:")
print(f"      Mean: {normalized_segments.mean():.4f}")
print(f"      Std: {normalized_segments.std():.4f}")
print(f"      Min: {normalized_segments.min():.4f}")
print(f"      Max: {normalized_segments.max():.4f}")
print(f"   ✓ Normalizacija uspješna!")
```

```python
#Provera kvaliteta podataka-da li su svi uzorci brojevi, da li ima beskonačnih vrednosti ili NaN-ova, da li su svi segmenti nula itd.

def check_data_quality(segments):
    """ 
    Proveri kvalitetu podataka
    """
    n_segments,n_channels,n_samples=segments.shape

    #brojaci za razlicite vrste segmenata
    quality_report={
        'total_segments':n_segments,
        'nan_segments':0, #treba da proverimo nan
        'inf_segments':0, #treba da proverimo inf
        'zero_segmnets':0, #treba da proverimo da li su svi uzorci nula
        'valid_segments':0 #ako nije ni nan ni inf ni nula, onda je validan segment
    }

    for i in range(n_segments):
        has_nan=np.any(np.isnan(segments[i]))
        has_inf=np.any(np.isinf(segments[i]))
        is_zero=np.allclose(segments[i],0)

        if has_nan:
            quality_report['nan_segments']+=1
        elif has_inf:
            quality_report['inf_segments']+=1
        elif is_zero:
            quality_report['zero_segmnets']+=1
        else:
            quality_report['valid_segments']+=1
    
    return quality_report

#Test provere kvaliteta
print("Provera kvaliteta podataka...")
quality=check_data_quality(normalized_segments)

print(f"   Ukupno segmenata: {quality['total_segments']}")
print(f"   Validni segmenti: {quality['valid_segments']} ({100*quality['valid_segments']/quality['total_segments']:.1f}%)")
print(f"   Segmenti sa NaN: {quality['nan_segments']}")
print(f"   Segmenti sa Inf: {quality['inf_segments']}")
print(f"   Segmenti sa samo nulama: {quality['zero_segmnets']}")

if quality['valid_segments']==quality['total_segments']:
    print("Svi segmenti su validni!")
```

```python
def process_edf_file_complete(edf_file, filter_low, filter_high, window_len_sec, overlap):
    """
    Kompletan pipeline obrade jednog EDF fajla
    """
    try:
        # 1. Učitaj EDF
        if edf_file.suffix == ".edf":
            raw = mne.io.read_raw_edf(edf_file, preload=True)
        elif edf_file.suffix == ".fif":
            raw = mne.io.read_raw_fif(edf_file, preload=True)
        signal_data = raw.get_data()  # (channels, samples)
        TARGET_CHANNELS = 23

        if signal_data.shape[0] > TARGET_CHANNELS:
            signal_data = signal_data[:TARGET_CHANNELS, :]
        elif signal_data.shape[0] < TARGET_CHANNELS:
            print(f"⚠️ Preskačem {edf_file.name} - ima {signal_data.shape[0]} kanala")
            return None, None, None
        fs = int(raw.info['sfreq'])
        
        # 2. Primijeni filter na sve kanale
        filtered_data = np.zeros_like(signal_data)
        for ch in range(signal_data.shape[0]):
            filtered_data[ch, :] = apply_bandpass_filter(
                signal_data[ch, :], fs, filter_low, filter_high
            )
        
        # 3. Segmentacija
        segments = segment_signal(filtered_data, fs, window_len_sec, overlap)
        
        # 4. Normalizacija
        normalized = normalize_segments(segments)
        
        # 5. Provjera kvalitete
        quality = check_data_quality(normalized)
        
        return normalized, quality, fs
    
    except Exception as e:
        print(f"❌ Greška: {str(e)}")
        return None, None, None

# ════════════════════════════════════════════════════════════════════════════
# OBRADA SVE 100 PACIJENATA SA SEIZURE LABELAMA
# ════════════════════════════════════════════════════════════════════════════

print("\n" + "="*80)
print("🚀 OBRADA SVE 100+ PACIJENATA (ORIGINAL + AUGMENTOVANI)")
print("="*80 + "\n")

all_segments = []
all_labels = []
processing_stats = []

# Kreiraj listu svih EDF fajlova za sve pacijente
total_files = 0
for patient_id in ALL_PATIENTS:
    patient_dir = DATA_RAW_DIR / patient_id
    if patient_dir.exists():
        total_files += len(list(patient_dir.glob(f"{patient_id}_*.edf")))
        total_files += len(list(patient_dir.glob(f"{patient_id}_*.fif")))

print(f"📊 Ukupno fajlova za obradu: {total_files}\n")

file_count = 0
for patient_idx, patient_id in enumerate(ALL_PATIENTS, 1):
    patient_dir = DATA_RAW_DIR / patient_id
    
    if not patient_dir.exists():
        continue
    
    # Pronađi sve EDF fajlove za ovog pacijenta
    edf_files = sorted(
        list(patient_dir.glob(f"{patient_id}_*.edf")) +
        list(patient_dir.glob(f"{patient_id}_*.fif"))
    )
    
    if not edf_files:
        continue
    
    # Učitaj seizure anotacije za originalne pacijente
    seizure_annotations = {}
    if patient_id.startswith('chb') and not 'aug' in patient_id:
        # Originalni pacijent - učitaj anotacije
        seizure_annotations = load_seizure_annotations(patient_id)
        print(patient_id)
        print("Broj anotacija:", len(seizure_annotations))

        if len(seizure_annotations) > 0:
            print("Prvi zapis:", list(seizure_annotations.keys())[:3])
    
    for edf_file in edf_files:
        file_count += 1
        print(f"[{file_count}/{total_files}] Obrada: {edf_file.name}", end="", flush=True)
        
        segments, quality, fs = process_edf_file_complete(
            edf_file, FILTER_LOW, FILTER_HIGH, WINDOW_LENGTH, OVERLAP
        )
        
        if segments is not None:
            # ─── Odredi labele za segmente ───
            filename = edf_file.name
            seizures = seizure_annotations.get(filename, [])
            
            # Za svaki segment, odredi je li seizure (1) ili non-seizure (0)
            segment_labels = []
            for seg_idx in range(segments.shape[0]):
                is_seizure = False
                
                # Odredi vrijeme segmenta
                step_samples = int(WINDOW_LENGTH * fs * (1 - OVERLAP))
                start_sample = seg_idx * step_samples
                end_sample = start_sample + int(WINDOW_LENGTH * fs)
                start_time = start_sample / fs
                end_time = end_sample / fs
                
                # Provjeri preklap sa seizure vremenom
                for seizure in seizures:
                    seizure_start = seizure['start']
                    seizure_end = seizure.get('end', seizure_start + 60)
                    
                    if not (end_time < seizure_start or start_time > seizure_end):
                        is_seizure = True
                        break
                
                segment_labels.append(1 if is_seizure else 0)
            
            all_segments.append(segments)
            all_labels.extend(segment_labels)
            
            print(f" ✓ ({len(segment_labels)} seg)")
            processing_stats.append({
                'file': edf_file.name,
                'n_segments': segments.shape[0],
                'quality': quality,
                'n_seizure': sum(segment_labels),
                'n_nonseizure': len(segment_labels) - sum(segment_labels)
            })
        else:
            print(f" ❌")
```

```python
# Kombinuj sve segmente
print("\n" + "="*80)
print("📦 KOMBINOVANJE SVIH SEGMENATA + BALANSIRANJE LABELA (50/50)")
print("="*80 + "\n")

if all_segments:

    # Uzmi samo prvih N grupa segmenata da ne pukne memorija
    MAX_GROUPS = 20

    selected_segments = all_segments[:MAX_GROUPS]

    combined_segments = np.vstack(selected_segments)

    # Izračunaj koliko labela pripada ovim segmentima
    n_segments_used = sum(seg.shape[0] for seg in selected_segments)

    all_labels_array = np.array(
        all_labels[:n_segments_used],
        dtype=np.int32
    )

    print(f"📊 Prije balansiranja:")
    print(f"   Ukupni broj segmenata: {combined_segments.shape[0]:,}")
    print(f"   Dimenzije: {combined_segments.shape}")
    print(f"   Memorija: {combined_segments.nbytes / (1024**3):.2f} GB")
    print(f"   Seizure (1): {(all_labels_array == 1).sum():,} ({100*(all_labels_array == 1).sum()/len(all_labels_array):.1f}%)")
    print(f"   Non-seizure (0): {(all_labels_array == 0).sum():,} ({100*(all_labels_array == 0).sum()/len(all_labels_array):.1f}%)")

    print(f"\n⚖️  BALANSIRANJE LABELA NA 50/50...\n")

    n_seizure = (all_labels_array == 1).sum()
    n_nonseizure = (all_labels_array == 0).sum()

    min_class = min(n_seizure, n_nonseizure)

    seizure_indices = np.where(all_labels_array == 1)[0]
    nonseizure_indices = np.where(all_labels_array == 0)[0]

    np.random.seed(42)

    seizure_indices = np.random.choice(
        seizure_indices,
        min_class,
        replace=False
    )

    nonseizure_indices = np.random.choice(
        nonseizure_indices,
        min_class,
        replace=False
    )

    balanced_indices = np.concatenate([
        seizure_indices,
        nonseizure_indices
    ])

    np.random.shuffle(balanced_indices)

    combined_segments_balanced = combined_segments[balanced_indices]
    labels_balanced = all_labels_array[balanced_indices]

    print(f"📊 Poslije balansiranja:")
    print(f"   Ukupni broj segmenata: {combined_segments_balanced.shape[0]:,}")
    print(f"   Seizure (1): {(labels_balanced == 1).sum():,} ({100*(labels_balanced == 1).sum()/len(labels_balanced):.1f}%)")
    print(f"   Non-seizure (0): {(labels_balanced == 0).sum():,} ({100*(labels_balanced == 0).sum()/len(labels_balanced):.1f}%)")

else:
    print("⚠️  Nema uspješno obrađenih fajlova!")
```

```python
#Ucitavanje dataseta za LSTM
import numpy as np

print("="*80)
print("🧠 UČITAVANJE PODATAKA ZA LSTM")
print("="*80)

X_train = np.load(DATA_PROCESSED_DIR / "X_train.npy")
y_train = np.load(DATA_PROCESSED_DIR / "y_train.npy")

X_val = np.load(DATA_PROCESSED_DIR / "X_val.npy")
y_val = np.load(DATA_PROCESSED_DIR / "y_val.npy")

X_test = np.load(DATA_PROCESSED_DIR / "X_test.npy")
y_test = np.load(DATA_PROCESSED_DIR / "y_test.npy")

print("Original shape:")
print("Train:", X_train.shape)
print("Val:", X_val.shape)
print("Test:", X_test.shape)

# LSTM očekuje (samples, timesteps, features)
X_train = np.transpose(X_train, (0, 2, 1))
X_val = np.transpose(X_val, (0, 2, 1))
X_test = np.transpose(X_test, (0, 2, 1))

print("\nLSTM shape:")
print("Train:", X_train.shape)
print("Val:", X_val.shape)
print("Test:", X_test.shape)
```

```python
# ============================================================================
# PYTORCH LSTM MODEL
# ============================================================================

import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

print("="*80)
print("🧠 PRIPREMA PODATAKA ZA PYTORCH")
print("="*80)

# numpy -> torch
X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.float32)

X_val_t = torch.tensor(X_val, dtype=torch.float32)
y_val_t = torch.tensor(y_val, dtype=torch.float32)

X_test_t = torch.tensor(X_test, dtype=torch.float32)
y_test_t = torch.tensor(y_test, dtype=torch.float32)

train_dataset = TensorDataset(X_train_t, y_train_t)
val_dataset = TensorDataset(X_val_t, y_val_t)
test_dataset = TensorDataset(X_test_t, y_test_t)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)

print("✅ Dataset spreman")

# ============================================================================
# LSTM MODEL
# ============================================================================

class SeizureLSTM(nn.Module):

    def __init__(self):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size=23,
            hidden_size=32,
            batch_first=True
        )

        self.fc1 = nn.Linear(32, 256)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(256, 1)

    def forward(self, x):

        lstm_out, (hidden, cell) = self.lstm(x)

        x = hidden[-1]

        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)

        x = self.fc2(x)

        return x.squeeze()

model = SeizureLSTM()

print(model)
```

```python
# ============================================================================
# TRENIRANJE LSTM MODELA
# ============================================================================

import torch.optim as optim

criterion = nn.BCEWithLogitsLoss()

optimizer = optim.Adam(
    model.parameters(),
    lr=0.001
)

EPOCHS = 25

train_losses = []
val_losses = []

print("="*80)
print("🚀 TRENIRANJE LSTM MODELA")
print("="*80)

for epoch in range(EPOCHS):

    # TRAIN
    model.train()

    running_loss = 0.0

    for X_batch, y_batch in train_loader:

        optimizer.zero_grad()

        outputs = model(X_batch)

        loss = criterion(outputs, y_batch)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    # VALIDATION
    model.eval()

    val_loss = 0.0

    with torch.no_grad():

        for X_batch, y_batch in val_loader:

            outputs = model(X_batch)

            loss = criterion(outputs, y_batch)

            val_loss += loss.item()

    val_loss = val_loss / len(val_loader)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    print(
        f"Epoch [{epoch+1}/{EPOCHS}] | "
        f"Train Loss: {train_loss:.4f} | "
        f"Val Loss: {val_loss:.4f}"
    )

print("\n✅ Treniranje završeno!")
```

```python
# ============================================================================
# FEATURE EXTRACTION IZ EEG SEGMENATA
# ============================================================================

import numpy as np

def extract_features(X):

    features = []

    for segment in X:

        segment_features = []

        for ch in range(segment.shape[1]):

            signal = segment[:, ch]

            segment_features.extend([
                np.mean(signal),
                np.std(signal),
                np.max(signal),
                np.min(signal),
                np.var(signal),
                np.sqrt(np.mean(signal**2))  # RMS
            ])

        features.append(segment_features)

    return np.array(features)

print("🔄 Ekstrakcija feature-a...")

X_train_feat = extract_features(X_train)
X_val_feat = extract_features(X_val)
X_test_feat = extract_features(X_test)

print("✅ Feature extraction završena")

print("Train:", X_train_feat.shape)
print("Val:", X_val_feat.shape)
print("Test:", X_test_feat.shape)
```

```python
# ============================================================================
# GRU MODEL
# ============================================================================

import torch
import torch.nn as nn

class SeizureGRU(nn.Module):

    def __init__(self):
        super().__init__()

        self.gru = nn.GRU(
            input_size=23,
            hidden_size=64,
            batch_first=True
        )

        self.fc1 = nn.Linear(64, 32)

        self.relu = nn.ReLU()

        self.dropout = nn.Dropout(0.3)

        self.fc2 = nn.Linear(32, 1)

    def forward(self, x):

        gru_out, hidden = self.gru(x)

        x = hidden[-1]

        x = self.fc1(x)

        x = self.relu(x)

        x = self.dropout(x)

        x = self.fc2(x)

        return x.squeeze()

gru_model = SeizureGRU()

print(gru_model)
```

```python
# ============================================================================
# TRENIRANJE GRU MODELA
# ============================================================================

import torch.optim as optim

criterion = nn.BCEWithLogitsLoss()

optimizer = optim.Adam(
    gru_model.parameters(),
    lr=0.001
)

EPOCHS = 10

gru_train_losses = []
gru_val_losses = []

print("="*80)
print("🚀 TRENIRANJE GRU MODELA")
print("="*80)

for epoch in range(EPOCHS):

    gru_model.train()

    running_loss = 0

    for X_batch, y_batch in train_loader:

        optimizer.zero_grad()

        outputs = gru_model(X_batch)

        loss = criterion(outputs, y_batch)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

    train_loss = running_loss / len(train_loader)

    gru_model.eval()

    val_loss = 0

    with torch.no_grad():

        for X_batch, y_batch in val_loader:

            outputs = gru_model(X_batch)

            loss = criterion(outputs, y_batch)

            val_loss += loss.item()

    val_loss /= len(val_loader)

    gru_train_losses.append(train_loss)
    gru_val_losses.append(val_loss)

    print(
        f"Epoch [{epoch+1}/{EPOCHS}] | "
        f"Train Loss: {train_loss:.4f} | "
        f"Val Loss: {val_loss:.4f}"
    )

print("\n✅ GRU trening završen!")
```
